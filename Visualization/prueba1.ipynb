```python
#import the libraries
import pandas as pd
import numpy as np
import panel as pn
pn.extension('tabulator') #allows displaying and editing a pandas DataFrame

import hvplot.pandas
```

```python
data_frame = pd.read_csv("cansat.csv")
```

```python
data_frame
```

```python
columns = ["Orientation", "Distance", "Deployment", "Velocity"] #the only columns we want to display in this widget
columns_df = data_frame[columns]
#clean_columns = columns_df.style.hide(axis="index -- would show the data_frame without indexes. Doesnt work as an argument
```

```python
chart_data = pn.widgets.DataFrame(columns_df, name='CSV Data', width=500) #creates the widget with desired columns
type(chart_data)
```

```python
chart_data
```

```python
UV_time_columns = ["UV", "Time"] #select columns for this graph
UV_time_df = data_frame[UV_time_columns] #create new dataframe with only those columns
UVradiation_plot = UV_time_df.hvplot(x="Time", y="UV", title="UV Radiation over Time") #since we are using hvplot we needed to make the new dataframe
```

```python
UVradiation_plot
```

```python
UVradiation_panel = UVradiation_plot.opts(width=700) 
'''since a normal graph has an object type of curve, instead of hvplot.interactive.Interactive--
we cannot use the .panel() method in the template at the very bottom.
This line converts it from curve to holoviews.element.chart.Curve. We then add it to the template 
without the .panel() method.
''' 
```

```python
Pressure_altitude_columns = ["Altitude", "Pressure"]
Pressure_altitude_df = data_frame[Pressure_altitude_columns]
Pressure_altitude_plot = Pressure_altitude_df.hvplot(x="Altitude", y="Pressure", title="Altitude over Pressure")
```

```python
Pressure_altitude_plot
```

```python
Pressure_altitude_panel = Pressure_altitude_plot.opts(width=600) 
```

```python
Velocity_time_columns = ["Time", "Velocity"]
Velocity_time_df = data_frame[Velocity_time_columns]
Velocity_time_plot = Velocity_time_df.hvplot(x="Time", y="Velocity", title="Velocity over Time")
```

```python
Velocity_time_plot
```

```python
Velocity_time_panel = Velocity_time_plot.opts(width=600) 
```

```python
Altitude_time_columns = ["Time", "Altitude"]
Altitude_time_df = data_frame[Altitude_time_columns]
Altitude_time_plot = Altitude_time_df.hvplot(x="Time", y="Altitude", title="Altitude over Time")
```

```python
Altitude_time_plot
```

```python
Altitude_time_panel = Altitude_time_plot.opts(width=600) 
```

```python
#Layout using Template
template = pn.template.FastListTemplate(
    title='Panel Cansat', 
    main=[
        pn.Row(
            pn.Column(UVradiation_panel, margin=(0, 25)), 
            chart_data
        ), 
        pn.Row(
            pn.Column(Pressure_altitude_plot, margin=(0, 25)), 
            pn.Column(Velocity_time_plot)
        ),
        pn.Row(
            pn.Column(Altitude_time_plot)
        )
    ],
    accent_base_color="#0000FF",
    header_background="#0000FF"
)
template.servable();
```
